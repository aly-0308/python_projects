```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

kernels = {
    "edge detection": np.array([[-1, -1, -1],
                                 [-1,  8, -1],
                                 [-1, -1, -1]]),
    "blurring": np.array([[1/9, 1/9, 1/9],
                           [1/9, 1/9, 1/9],
                           [1/9, 1/9, 1/9]]),
    "sharpening": np.array([[0, 1, 0],
                            [1, -4, 1],
                            [0, 1, 0]]),
    "nothing": np.array([[0, 0, 0],
                         [0, 1, 0],
                         [0, 0, 0]])
}

def apply_kernel(image_path, kernel_type):
    # Read the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    #  cv2.IMREAD_GRAYSCALE making sure the image has been read in gray scale

    # Apply the convolution operation
    result = cv2.filter2D(img, -1, kernels[kernel_type])
    # '-1' = ddepth: desired depth of the destination image. If it is negative, it will be the same as src.depth().
    # Other options cv2.CV_8U for 8-bit unsigned integers, cv2.CV_32F for 32-bit floating point

    # Display the original and processed images
    plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original Image')
    plt.subplot(122), plt.imshow(result, cmap='gray'), plt.title(kernel_type.capitalize())
    plt.show()

if __name__ == "__main__":
    image_path = 'road.jpg'
    apply_kernel(image_path, "edge detection")
    image_path = 'dog.jpg'
    apply_kernel(image_path, "blurring")
```
